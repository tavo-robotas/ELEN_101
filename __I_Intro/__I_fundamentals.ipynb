## Atomas

Atomas yra dalel, kuri sudaro <code>**proton ir neutron branduolys, apsuptas elektron debesies**</code>. Atomas yra pagrindin chemini element dalel, o cheminiai elementai skiriasi vienas nuo kito pagal j atomuose esani proton skaii. Pavyzdiui, bet kuris atomas, kuriame yra 11 proton, yra natris, o bet kuris atomas, kuriame yra 29 protonai, yra varis. Atomas su 1 protonu yra vandenilis, pats lengviausias ir populiariausas atomas visatoje.<br>

Atomai yra tokie mai kad tas dydis yra maesnis u trumpiausi matomos viesos bangos ilg, o tai reikia, kad mons nemato atom prastais mikroskopais. Atomai yra tokie mai, kad net tiksliai numatyti j elgesio naudojant klasikin fizik nemanoma dl kvantini efekt.<br>

<code>**Atomo sudtis yra i subatomini daleli: proton, neutron, elektron.**</code><br>
Protonai turi teigiam elektros krv ( + ), elektronai neigiam (  ), o neutronai neturi elektros krvio. <code>**Jei proton ir elektron skaiius yra lygus, atomas yra elektrikai neutralus**</code>. Jei atomas turi daugiau ar maiau elektron nei proton, tai jis turi atitinkamai bendr neigiam arba teigiam krv  tokie atomai vadinami jonais.

## Elektronas

Angl mokslininkas fizikas vardu. J. J. Thomson 1897 m. atrado, kad katodiniai spinduliai nra elektromagnetins bangos, o sudaryti i daleli, nes jas gali nukreipti elektriniai ir magnetiniai laukai. Jis nustat, kad ios dalels yra <code>**1800 kart lengvesns u vandenil (atsimenante - lengviausias ir populiariausias atomas visatoje!)**</code>. Thomsonas padar ivad, kad ios dalels kilo i katodo atom - tai buvo <code>**subatomins dalels**</code>. Jis ias naujas daleles pavadino kneliais, bet vliau jos buvo pervadintos  elektronus. Thomsonas taip pat parod, kad elektronai buvo identiki dalelms, kurias iskiria fotoelektrins ir radioaktyvios mediagos. Greitai buvo pripainta, kad elektronai yra dalels, perneanios elektros srov metaliniuose laiduose.Thomsonas padar ivad, kad ie elektronai atsirado i pai katodo atom jo instrumentuose, o tai reikia, kad atomai nra nedalomi, kaip man kitas mokslininkas vardu Daltonas.
