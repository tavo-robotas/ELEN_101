## Atomas

Atomas yra dalelė, kurią sudaro <code>**protonų ir neutronų branduolys, apsuptas elektronų debesies**</code>. Atomas yra pagrindinė cheminių elementų dalelė, o cheminiai elementai skiriasi vienas nuo kito pagal jų atomuose esančių protonų skaičių. Pavyzdžiui, bet kuris atomas, kuriame yra 11 protonų, yra natris, o bet kuris atomas, kuriame yra 29 protonai, yra varis. Atomas su 1 protonu yra vandenilis, pats lengviausias ir populiariausas atomas visatoje.<br>

Atomai yra tokie maži kad tas dydis yra mažesnis už trumpiausią matomos šviesos bangos ilgį, o tai reiškia, kad žmonės nemato atomų įprastais mikroskopais. Atomai yra tokie maži, kad net tiksliai numatyti jų elgesio naudojant klasikinę fiziką neįmanoma dėl kvantinių efektų.<br>
<br>

<code>**Atomo sudėtis yra iš subatominių dalelių: protonų, neutronų, elektronų.**</code><br>
Protonai turi teigiamą elektros krūvį ( + ), elektronai neigiamą ( – ), o neutronai neturi elektros krūvio. <code>**Jei protonų ir elektronų skaičius yra lygus, atomas yra elektriškai neutralus**</code>. Jei atomas turi daugiau ar mažiau elektronų nei protonų, tai jis turi atitinkamai bendrą neigiamą arba teigiamą krūvį – tokie atomai vadinami jonais.

## Elektronas

Anglų mokslininkas fizikas vardu. J. J. Thomson 1897 m. atrado, kad katodiniai spinduliai nėra elektromagnetinės bangos, o sudaryti iš dalelių, nes jas gali nukreipti elektriniai ir magnetiniai laukai. Jis nustatė, kad šios dalelės yra <code>**1800 kartų lengvesnės už vandenilį (atsimenante - lengviausias ir populiariausias atomas visatoje!)**</code>. Thomsonas padarė išvadą, kad šios dalelės kilo iš katodo atomų - tai buvo <code>**subatominės dalelės**</code>. Jis šias naujas daleles pavadino kūneliais, bet vėliau jos buvo pervadintos į elektronus. Thomsonas taip pat parodė, kad elektronai buvo identiški dalelėms, kurias išskiria fotoelektrinės ir radioaktyvios medžiagos. Greitai buvo pripažinta, kad elektronai yra dalelės, pernešančios elektros srovę metaliniuose laiduose.Thomsonas padarė išvadą, kad šie elektronai atsirado iš pačių katodo atomų jo instrumentuose, o tai reiškia, kad atomai nėra nedalomi, kaip manė kitas mokslininkas vardu Daltonas. 

<div class='alert alert-info'>
Beje graikiškas žodis, reiškiantis gintarą <b>(ἤλεκτρον - ḗlektron)</b> buvo žodžio elektronas ir elektra šaltinis. Dėl to kad nuo labai seno buvo žinoma kad kad kai kurios medžiagos, tarkime kaip tas gintaras po trynimo į paviršių pritraukia lengvas daleles.

saugoma baterijoje), galima padaryti kad elektronai pradėtu kryptingai judėti iš vieno atomo į kitą!
